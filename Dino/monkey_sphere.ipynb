```python
import cf_ownership as cfo
import counterfeit_utils as cfu
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
slug = 'boredapeyachtclub'
look_sims = cfu.get_look_sims(slug,remove_ders=True)
sales_df = cfo.get_ownership_sales(slug,tuple(look_sims))
sales_df['timestamp'] = pd.to_datetime(sales_df['timestamp'],unit='s')
#make a day column
sales_df['day'] = sales_df['timestamp'].dt.date
```

```python
#group by timestamp and plot sale prices
no_bored_df = sales_df.query('slug != "boredapeyachtclub"')
bored_df = sales_df.query('slug == "boredapeyachtclub"')
#Convert timestamp to datetime
no_bored_df['timestamp'] = pd.to_datetime(no_bored_df['timestamp'],unit='s')
#make a month column that gets year and month but no day
no_bored_df['day'] = no_bored_df['timestamp'].dt.date
no_bored_df['log_sale_price'] = no_bored_df['sale_price'].apply(lambda x: np.log(x))
no_bored_df = no_bored_df.groupby('day')['sale_price'].mean() 
# no_bored_df.plot(y='log_sale_price',title='Average Sale Price Over Time')
bored_df['timestamp'] = pd.to_datetime(bored_df['timestamp'],unit='s')
#make a day column
bored_df['day'] = bored_df['timestamp'].dt.date
bored_df['log_sale_price'] = bored_df['sale_price'].apply(lambda x: np.log(x))
bored_df = bored_df.groupby('day')['sale_price'].mean() 
# bored_df.plot(y='log_sale_price',title='Average Sale Price Over Time')
#merge the two dataframes and compute ratio of bored to non bored
merged_df = pd.merge(no_bored_df,bored_df,how='inner',on='day')
merged_df['ratio'] = merged_df['sale_price_x']/merged_df['sale_price_y']
merged_df['ratio'].plot(title='Ratio of Bored to Non-Bored Sale Prices Over Time')
plt.ylim(0,0.01)
```
