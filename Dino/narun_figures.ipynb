```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

data = synth.original_data
synth_outcome = data.synth_outcome
time = data.dataset[data.time].unique()
time = [datetime.strptime(t, '%Y-%m-%d') for t in time]

# Determine appropriate limits for y-axis
max_value = max(np.max(data.treated_outcome_all),
                np.max(data.synth_outcome))
min_value = min(np.min(data.treated_outcome_all),
                np.min(data.synth_outcome))

synth_label = "Synthetic Azuki"
treated_label = 'Azuki'
treatment_label = 'Beanz Official'

fig = plt.figure(figsize=(15, 8))

# Add vertical grid lines with respect to the treatment date
treatment_date = datetime.strptime('2022-03-28', '%Y-%m-%d')
interval = timedelta(days=10)  # Set the interval for grid lines
start_date = time[0]
end_date = time[-1]

current_date = treatment_date
while current_date >= start_date:
    plt.axvline(current_date, linestyle='--', color='lightgray', linewidth=0.5)
    current_date -= interval

current_date = treatment_date + interval
while current_date <= end_date:
    plt.axvline(current_date, linestyle='--', color='lightgray', linewidth=0.5)
    current_date += interval

# Add a more prominent vertical line for the treatment date
plt.axvline(treatment_date, linestyle='--', color='gray', linewidth=2)



# Make plot
plt.plot(time, synth_outcome.T, 'r--', label=synth_label)
plt.plot(time, data.treated_outcome_all, 'b-', label=treated_label)

# Set y-axis limits
plt.ylim(0, 1.2 * abs(max_value))  # Do abs() in case min is positive, or max is negative
plt.xlim(time[0], time[-1])

# Annotate the treatment label
plt.annotate(treatment_label, 
    xy=(treatment_date, data.treated_outcome[-1] * (1 + 0.2 * np.sign(data.treated_outcome[-1] - data.treated_outcome[0]))),
    xytext=(-250, -4),
    xycoords='data',
    textcoords='offset points',
    fontsize=25,
    arrowprops=dict(arrowstyle="->"))

# Customize the plot
plt.grid(visible=False)
plt.legend(loc='upper left', fontsize=25)
plt.ylabel('Sale Price', fontdict={'fontsize': 25})
plt.xlabel('Date', fontdict={'fontsize': 25})

# Set x-ticks with respect to the treatment date
xticks = []
current_date = treatment_date
while current_date >= start_date:
    xticks.append(current_date)
    current_date -= interval

current_date = treatment_date + interval
while current_date <= end_date:
    xticks.append(current_date)
    current_date += interval

xticks = sorted(xticks)
plt.xticks(xticks, [date.strftime('%Y-%m-%d') for date in xticks])

# Set y-ticks every 5 units
yticks = np.arange(0, 1.2 * abs(max_value), 5)
plt.yticks(yticks)

# Add horizontal grid lines
for ytick in yticks:
    plt.axhline(ytick, linestyle='--', color='lightgray', linewidth=0.5)

plt.savefig('plots/azuki_synth_plot.pdf', format='pdf', bbox_inches='tight')
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Turn wallet sales from a list of series into one big list and then plot
#iterate through fils in ownership_out directory
import os
import pickle

import matplotlib as mpl
# from matplotlib import font_manager

mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('narunraman.mplstyle')

# Define the directory containing the pickle files
directory = 'ownership_out'

# Step 1: List all files in the directory
all_files = os.listdir(directory)

# Step 2: Filter the list to include only pickle files
pickle_files = [f for f in all_files if f.endswith('.pkl') or f.endswith('.pickle')]
max_values = []
# Step 3: Iterate through the list of pickle files and open each one
for pickle_file in pickle_files:
    file_path = os.path.join(directory, pickle_file)
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
        slug,wallet_sales_1 = data
        max_values += [series.max() for series in wallet_sales_1]
combined_series = pd.Series(max_values)
percentile_99 = combined_series.quantile(0.99)

# Step 3: Filter out values above the 99th percentile
filtered_series = combined_series[combined_series <= percentile_99]
# Step 2: Convert the combined Series to a list
combined_list = filtered_series.tolist()
plt.figure(figsize=(10, 6))
print(len(set(combined_list)))
n, bins, patches = plt.hist(combined_list, bins=10000, density=True, cumulative=True, histtype='step', label='Look Sim first', linewidth=3)
plt.xlim(0, 15)
plt.xlabel('Price Ratio', fontsize = 25, fontweight='normal')
plt.ylabel('CDF', fontsize = 25, fontweight='normal')
# plt.legend()
# plt.title('Distribution of Combined Series')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Find the y-values where the vertical lines intersect the histogram
y1 = n[np.searchsorted(bins, 1)]
y5 = n[np.searchsorted(bins, 5)]

# # Add horizontal lines at the intersection points
# plt.axhline(y=y1, color='blue', linestyle='--', linewidth=2)
# plt.axhline(y=y5, color='blue', linestyle='--', linewidth=2)

# Add vertical lines at x = 1 and x = 5, stopping at the histogram
plt.plot([1, 1], [0, y1], color='red', linestyle='--', linewidth=2)
plt.plot([5, 5], [0, y5], color='red', linestyle='--', linewidth=2)

# Add horizontal lines at the intersection points, stopping at the vertical lines
plt.plot([0, 1], [y1, y1], color='blue', linestyle='--', linewidth=2)
plt.plot([0, 5], [y5, y5], color='blue', linestyle='--', linewidth=2)


plt.savefig('plots/ownership_out.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
